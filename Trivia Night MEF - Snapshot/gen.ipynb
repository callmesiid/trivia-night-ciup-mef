```python
import os
import json
import sys

# --- CONFIGURATION ---
# Le dossier où sont tes images
IMAGE_FOLDER = "Mix & Twist"
# Le fichier JS à créer
OUTPUT_FILE = "mix_data.js"

try:
    CURRENT_DIR = os.path.dirname(os.path.abspath(__file__))
except NameError:
    CURRENT_DIR = os.getcwd()

FULL_IMG_PATH = os.path.join(CURRENT_DIR, IMAGE_FOLDER)

print("------------------------------------------------")
print(f"   GÉNÉRATEUR SNAPSHOT (Mix & Twist)   ")
print("------------------------------------------------")

if not os.path.exists(FULL_IMG_PATH):
    print(f"/!\\ ERREUR : Le dossier '{IMAGE_FOLDER}' est introuvable !")
    sys.exit()

images_list = []

# Extensions acceptées
valid_exts = ('.png', '.jpg', '.jpeg', '.webp')

for f in os.listdir(FULL_IMG_PATH):
    if f.lower().endswith(valid_exts):
        # On nettoie le nom pour la Régie
        # Exemple : "zinedine_zidane.jpg" -> "Zinedine Zidane"
        raw_name = os.path.splitext(f)[0]
        clean_name = raw_name.replace('_', ' ').replace('-', ' ').title()
        
        # Chemin relatif pour le HTML
        # Attention : on force les slashs / pour le web, pas les backslashs \
        rel_path = f"{IMAGE_FOLDER}/{f}"
        
        images_list.append({
            "file": rel_path,
            "name": clean_name
        })

print(f"-> {len(images_list)} images trouvées.")

# Écriture du fichier JS
js_content = f"const MIX_PHOTOS = {json.dumps(images_list, indent=4, ensure_ascii=False)};"

try:
    with open(os.path.join(CURRENT_DIR, OUTPUT_FILE), "w", encoding="utf-8") as f:
        f.write(js_content)
    print(f"SUCCÈS ! Fichier '{OUTPUT_FILE}' généré.")
except Exception as e:
    print(f"Erreur d'écriture : {e}")
```
